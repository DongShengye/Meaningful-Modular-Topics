```python
import numpy as np
import matplotlib.pyplot as plt
def qr_linear_regression(X,y):
    m= X.shape[0]
  
    array1 = []
    for i in range(m):
        array1.append([1])
    array1 = np.array(array1)
    X = np.hstack((X,array1))
    
    Q, R = np.linalg.qr(X)
    b = np.linalg.solve(R, Q.T @ y)
    return b

X = np.array([[18.1],[8],[16.8],[2.8],[3.9],[12.7],[11.5],[9.5],[8],[13.2]])

y = np.array([[83],[42],[79],[16],[24],[73],[60],[44],[47],[67]])
plt.scatter(X, y)
b = qr_linear_regression(X,y)

plt.axline((0, b[1][0]), slope=b[0][0], color='C0', label='by slope')
plt.xlabel("X (inches)")
plt.ylabel("Y (years)")
plt.title("Tree age vs circumference of trunk")
```
